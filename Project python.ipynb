# 2. Data Preparation

### 2.1 Data source

# 3. Train

### 3.1 AI Model
โมเดลที่เราเลือกใช้ในการทำ Computer vision classification นั้นเป็น Yolo v8 เนื่องจากตัวของโมเดลนั้นสามารถใช้งานได้ง่ายมีการทำเครือข่ายประสาทเทียมภายในให้เรียบร้อยเราเพียงแค่นำข้อมูลภาพของเรานั้นไปเทรนให้กับโมเดลซึ่งความแม่นยำและความเสถียรนั้นสามารถใช้งานได้ดีและแม่นยำเราจึงเลือกใช้โมเดลนี้

### 3.2 Hyperparameters
Hyperparameter ที่เรามีการเซ็ทค่านั้นคือ epoch หรือจำนวนการเทรนของโมเดลซึ่งเราได้เซ็ทให้เทรนทั้งหมด 1000 ครั้ง ซึ่งเป็นค่าที่เราได้ลองปรับค่าจาก 100 ครั้งซึ่งเมื่อลองนำโมเดลออกมาใช้พบว่าความแม่นยำในการตรวจจับนั้นต่ำมากจนไม่สามารถใช้งานได้ จึงจำเป็นที่จะต้องใช้จำนวนการเทรนขึ้นไปถึง 1000 ครั้ง และมีการกำหนด Batchsize  ของข้อมูลที่จะให้โมเดลเทรนต่อรอบซึ่งเราได้เซ็ทไว้ที่ 32 เพื่อให้จำนวนของข้อมูลต่อรอบที่เทรนนั้นมีเยอะขึ้นแต่ไม่เกิน memory ของเครื่องที่ใช้งานได้

### 4.2 การอธิบายผลลัพธ์
จากผลลัพท์ข้างต้นที่เกิดขึ้นมานั้น
1. Confusion matrix: เป็นตารางที่บ่งบอกถึงสิ่งที่โมเดลทำนายออกมาเทียบกับความเป็นจริงซึ่งจะเห็นได้ว่าในตารางที่โมเดลทำนายออกมานั้นจะมีการทำนายผิดพลาดจากพื้นหลังกลายเป็นหัว หมวกกันน็อค และมอเตอร์ไซต์ซึ่งแสดงให้เห็นว่าโมเดลอาจเกิดข้อผิดพลาดในคานี้ได้เนื่องจากในขั้นตอนการ Label นั้นมีบางวัตถุที่เห็นได้ไม่ชัดเจนจึงไม่ได้ทำการ Label เข้าไปเมื่อโมเดลนำภาพนั้นมาทดสอบอาจจะทำให้ตรวจจับวัตถุนั้นเจอ แต่หากเทียบการทำนายที่ผิดพลาดของ False negative นั้นจะพบว่าในส่วนของหัวและหมวกกันน็อคเกิดความสับสนขึ้นเพียงเล็กน้อยเท่านั้น
2. Precision: คือค่าความแม่นยำที่โมเดลสามารถทำนายได้ว่าเป็นวัตถุชนิดนั้นได้ถูกต้องเป็นเท่าไร โดยที่โมเดลของเรานั้นมีค่า Precision ที่ 81% ซึ่งค่อนข้างสูงแต่ยังไม่เพียงพอ
3. Recall: คือความถูกต้องว่าโมเดลมีโอกาสที่จะทำนายวัตถุนั้นได้ถูกต้องตามความจริงเป็นเท่าไร โดยที่โมเดลของเรานั้นมีค่า Recall ที่ 88% ซึ่งค่อนข้างสูง
4. F1-Score: คือค่าเฉลี่ยของทั้ง Precision และ Recall (คล้าย Accuracy) โดยที่โมเดลของเรานั้นได้ค่าเป็น 84% ซึ่งถือว่าสูงแต่ยังสามารถเพิ่มความแม่นยำได้อีก
5. mAP50: คือการเปรียบเทียบ Overlayว่าการซ้อนทับของ label จริง เทียบกับที่โมเดลทำนายนั้นซ้อนทับกันขนาดไหนซึ่ง โมเดลของเรานั้นมีค่า mAP50 สูงถึง 0.965
